```python
import pandas as pd
```

```python
dataframe_exam = pd.read_csv(r'D:\rakamin_python\Data_Processing\job_data_analyst_details.csv')
```

```python
# pertanyaan satu
# antoni memiliki cita cita sebagai gamer, sehingga dia ingin bekerja di industry = video games yang memiliki rating perusahaan > 4.5,

# dimana lokasi perusahaan yang memenuhi kriteria tersebut di dataframe job_data_analyst_details.csv adalah sebagai berikut:

# kolom dalam dataframe_exam
# Index(['Unnamed: 0', 'Job Title', 'Salary Estimate', 'Job Description',
#        'Rating', 'Company Name', 'Location', 'Headquarters', 'Founded',
#        'Type of ownership', 'Industry', 'Revenue', 'Competitors',
#        'Easy Apply'],
#       dtype='object')

dataframe_exam[dataframe_exam['Industry'] == 'Video Games'][dataframe_exam['Rating'] > 4.5][['Location', 'Company Name', 'Rating']]

```

```python
# soal kedua
# berapa rata rata rating perusahaan dengan lokasi yang mengandung kata "Houston"

dataframe_exam[dataframe_exam['Location'].str.contains('Houston')]['Rating'].mean()
```

```python
# soal ketiga
# # dari perusahaan yang bergerak di industri = internet dan Founded > 2015, Job Title apa yang paling banyak lowongan pekerjaannya adalah?

dataframe_exam[(dataframe_exam['Industry'] == 'Internet') & (dataframe_exam['Founded'] > 2015)]['Job Title'].value_counts().idxmax()
```

```python
# soal keempat
# df1 didefinisikan sebagai perusahaan yang founded = 2018
# df2 didefinisikan sebagai perusahaan yang founded = 1995

# kemudian append kedua dataframe tersebut

# berapakah total lowongan yang disediakan setelah append tersebut?

df1 = dataframe_exam[dataframe_exam['Founded'] == 2018]
df2 = dataframe_exam[dataframe_exam['Founded'] == 1995]
df_append = pd.concat([df1, df2])  # Gunakan pd.concat() untuk menggabungkan DataFrame
total_job_titles = df_append['Job Title'].count()

print(total_job_titles)
```

```python
# soal kelima
# pada perusahaan yang founded > 2017 , rating >= 3.5 dan rating <= 4.5, jika dibuat kolom rating_baru yang bernilai rating+0.5

# maka berapa rata rata rating baru?

# Tambahkan kolom rating_baru
# Definisikan fungsi untuk menghitung rating_baru
def calculate_rating_baru(row):
    if (row['Founded'] > 2017) and (3.5 <= row['Rating'] <= 4.5):
        return row['Rating'] + 0.5
    else:
        return row['Rating']

# Tambahkan kolom rating_baru menggunakan fungsi biasa
dataframe_exam['rating_baru'] = dataframe_exam.apply(calculate_rating_baru, axis=1)

# Filter baris yang memenuhi kriteria
filtered_df = dataframe_exam[(dataframe_exam['Founded'] > 2017) & (3.5 <= dataframe_exam['Rating']) & (dataframe_exam['Rating'] <= 4.5)]

# Hitung rata-rata rating_baru hanya untuk baris yang memenuhi kriteria
average_new_rating = filtered_df['rating_baru'].mean()
print(average_new_rating)
```

```python
# soal keenam

# df1 didefinisikan semua jobs dengan kolom description mengandung kata attitude dan rating >= 1.0
# df2 didefinisikan semua jobs di new york,NY dan rating >= 1.0
# append kedua df tersebut, berapa mean dari rating dari seluruh jobs yang tersedia?

df1 = dataframe_exam[dataframe_exam['Job Description'].str.contains('attitude', case=False) & (dataframe_exam['Rating'] >= 1.0)]
df2 = dataframe_exam[dataframe_exam['Location'].str.contains('New York, NY') & (dataframe_exam['Rating'] >= 1.0)]
df_append = pd.concat([df1, df2])  # Gunakan pd.concat() untuk menggabungkan DataFrame
mean_rating = df_append['Rating'].mean()
print(mean_rating)
```

```python
# soal ketujuh
# ada berapa total lowongan pekerjaan di lokasi lake success, NY dan los Angeles, CA yang mensyaratkan menguasai SQL?

df_lake_success = dataframe_exam[(dataframe_exam['Location'] == 'Lake Success, NY') & (dataframe_exam['Job Description'].str.contains('SQL', case=False))]
df_los_angeles = dataframe_exam[(dataframe_exam['Location'] == 'Los Angeles, CA') & (dataframe_exam['Job Description'].str.contains('SQL', case=False))]   
total_jobs_sql = len(df_lake_success) + len(df_los_angeles)
print(total_jobs_sql)
print(len(df_lake_success))
print(len(df_los_angeles))
```

```python
# # soal kedelapan
# apa yang terjadi jika syntax how = ['inner','left','right'] pada saat merge tidak digunakan?
# contoh : df1.merge(df2, left_on= 'name', right_on = 'name')

# Jika syntax `how=['inner', 'left', 'right']` tidak digunakan pada saat melakukan merge, maka secara default pandas akan melakukan merge dengan metode 'inner join'.
# Artinya, hanya baris-baris yang memiliki nilai yang cocok di kedua DataFrame yang akan disertakan dalam hasil merge.
# Jika kita ingin mengubah jenis join yang digunakan, kita harus secara eksplisit menyebutkan jenis join tersebut dengan parameter `how`.
# Misalnya, jika kita ingin melakukan 'left join', kita harus menuliskan `how='left'`, yang akan mengembalikan semua baris dari DataFrame kiri (df1) dan hanya baris yang cocok dari DataFrame kanan (df2).
```
