# Import dataset

```python
import pandas as pd
df_customers = pd.read_excel('data_rakamin_customers.xlsx')
df_customers.head()
```

# 1. GroupBy Pandas

## a. Satu By Paramater

### a.1 Satu Kolom dan Satu Statistik

__Format Syntax__ : df.groupby(‘column_name’).agg({‘column’ : ‘statistics’})

```python
# CONTOH-1
# Ada berapa siswa (count registration_id) per province

df_customers.groupby('province').agg({'registration_id' : 'count'})
```

```python
# CONTOH-2
# Ada berapa UNIQUE siswa (count registration_id) per province

df_customers.groupby('province').agg({'registration_id' : 'nunique'})
```

```python
# CONTOH-3
# Ada rata-rata umur per province

df_customers.groupby('province').agg({'age' : 'mean'})
```

```python
# CONTOH-4
# Ada rata-rata umur per income_group

df_customers.groupby('income_group').agg({'age' : 'mean'})
```

```python
# CONTOH-5
# Berapa median website_visit_count per income_group?

df_customers.groupby('income_group').agg({'website_visit_count' : 'median'})
```

```python
df_customers.groupby("city").agg({
    "website_visit_count": ["mean", "max", "min"]
})
```

### a.2 Satu Kolom dan Multiple Statistik

__Format Syntax__ : df.groupby(‘column_name’).agg({‘column’ : [ ‘statistik_1’, ‘statistik_2’, . . . . . ] } )

```python
# CONTOH-1
# Berapa rata-rata dan median umur para customer per province?

df_customers.groupby('province').agg({'age' : ['mean','median']})
```

```python
# CONTOH-2
# Berapa rata-rata , median , termuda dan tertua umur per province?

df_customers.groupby('province').agg({'age' : ['mean','median','min','max']})
```

```python
# CONTOH-3
# Berapa rata-rata , dan maksimum age per website_cs_call_count?

df_customers.groupby('website_cs_call_count').agg({'age' : ['mean','max']})
```

### a.3 Multiple Kolom dan Multiple Statistik

```python
# CONTOH-1
## Hitunglah banyaknya customer, rata-rata & median umur, dan  
## website_purchase_rate minimum & maximum per income_group?


df_customers.groupby('income_group').agg({'registration_id' : ['count'],
                                          'age' : ['mean','median'],
                                          'website_purchase_rate' : ['min','max']})
```

```python
# CONTOH-2
## Hitunglah median website_recency, rata-rata website_purchase_amount, 
## dan rata-rata website_visit_count per city?

df_customers.groupby('city').agg({'website_recency' : ['median'],
                                  'website_purchase_amount' : ['mean'],
                                  'website_visit_count' : ['mean']})
```

## b. Multiple By Parameter

```python
# CONTOH-1
## Hitunglah banyaknya customer, rata-rata & median umur, 
## dan website_purchase_rate minimum & maximum per province dan income_group?

df_customers.groupby(['province','income_group']).agg({'registration_id' : ['nunique'],
                                                       'age' : ['mean','median'],
                                                       'website_purchase_rate' : ['min','max']})
```

```python
# CONTOH-2
## Hitunglah median website_recency, rata-rata website_purchase_amount, 
## dan rata-rata website_visit_count per province, city dan income_group?

df_customers.groupby(['province','city','income_group']).agg({'website_recency' : ['median'],
                                                              'website_purchase_amount' : ['mean'],
                                                              'website_visit_count' : ['mean']})
```

# 3. Pivot Table

```python
import pandas as pd
df_customers = pd.read_excel('data_rakamin_customers.xlsx')
```

## a. Sebagai aggregate function

```python
# CONTOH-1
## Hitunglah banyaknya customer, rata-rata & median umur, dan  website_purchase_rate minimum & maximum per income_group?

pd.pivot_table(df_customers,
              index = ['income_group'], # group by parameters
              values = ['registration_id','age','website_purchase_rate'], # values yang mau dihitung / aggregate
              aggfunc = {'registration_id' : ['nunique'],
                         'age' : ['mean','median'],
                         'website_purchase_rate' : ['min','max']}
              )
```

```python
# CONTOH-2 
## Hitunglah banyaknya customer, rata-rata & median umur, 
## dan website_purchase_rate minimum & maximum per province dan income_group?

pd.pivot_table(df_customers,
              index = ['province','income_group'],
              values = ['registration_id','age','website_purchase_rate'],
              aggfunc = {'registration_id' : ['nunique'],
                         'age' : ['mean','median'],
                         'website_purchase_rate' : ['min','max']}
              )
```

## b. Sebagai reshape table

```python
# CONTOH-1 : Kita buat dataframe sederhana supaya lebih mudah dilihat hasilnya

df = pd.DataFrame({'province' : ['DKI Jakarta','DKI Jakarta','Banten','Banten','Jawa Barat','Jawa Barat'],
                   'is_active' : [1,0,1,0,1,0],
                   'total_customers' : [40, 30, 60, 45, 20, 35]})
df
```

```python
pd.pivot_table(df,
              index = ['province'],
              columns = ['is_active'],
              values = ['total_customers']
              )
```

```python
# CONTOH 2
## Hitunglah jumlah customer yang accept campaign_1 dan tidak accept accept campagn_1 per income_group.

## accept campaign_1: ketika kolom accept_campaign_1 bernilai True
## tidak accept campaign_1:  ketika kolom accept_campaign_1 bernilai False


## Pada income_group berapa yang jumlah accept campaign lebih banyak dari yang tidak terima campaign?

### Kita lakukan proses groupby seperti sebelumnya

df_grouped = df_customers.groupby(['income_group','accept_campaign_1']).agg({'registration_id' : ['nunique']})
df_grouped = df_grouped.reset_index()
df_grouped.columns = ['income_group','accept_campaign_1','total_customers']
df_grouped
```

```python
# Kita akan ubah accept_campaign_1 menjadi 2 kolom

pd.pivot_table(df_grouped,
               index = ['income_group'],
               columns = ['accept_campaign_1'],
               values = ['total_customers']
              )
```

```python
# Supaya lebih mudah dalam analisis data, kita rapi kan hasil di atas dengan ubah nama kolomnya

df_pivot = pd.pivot_table(df_grouped,
                          index = ['income_group'],
                          columns = ['accept_campaign_1'],
                          values = ['total_customers']
                          )

df_pivot = df_pivot.reset_index()
df_pivot.columns = ['income_group','customer_non_accept','customer_accept']
df_pivot
```

```python
# Bisa juga kita lakukan filtering untuk mengambil baris yang accept > tidak accept

df_pivot[df_pivot['customer_accept']>df_pivot['customer_non_accept']]
```
